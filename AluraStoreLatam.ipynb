### Importación de datos



```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
calificacion_tienda3 = sum(tienda3['Calificación'])/len(tienda3['Calificación'])
print(f'Calificación promedio de la tienda 3: {calificacion_tienda3:.2f}')
```

```python
calificacion_tienda4 = sum(tienda4['Calificación'])/len(tienda4['Calificación'])
print(f'Calificación promedio de la tienda 4: {calificacion_tienda4:.2f}')
```

```python
'''
Apartado para realizar una gráfica comparando las calificaciones promedio de cada tienda

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ventas = [ventas_tienda1, ventas_tienda2, ventas_tienda3, ventas_tienda4]

'''
# Datos de ejemplo (reemplaza las variables con las calificaciones reales)
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion_tienda, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]

# Función para mostrar la calificación total en lugar de porcentajes
def calificacion_total(pct, allvals):
    absolute = pct / 100.*sum(allvals)  # Calcula el valor absoluto
    return f'{absolute:.2f}'  # Retorna el valor formateado

# Crear gráfica de tipo pie
plt.pie(calificaciones, labels=tiendas, autopct=lambda pct: calificacion_total(pct, calificaciones),
        colors=['mediumseagreen', 'skyblue', 'lightcoral', 'gold'], startangle=90)

# Título
plt.title('Distribución de Calificaciones por Tienda')

# Mostrar gráfica
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
#Productos más vendidos vs menos vendido por tienda
'''
En este paso, debes identificar qué productos fueron los más vendidos y los menos
vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos
destacaron en ventas en cada tienda.
'''
#----------------------------Tienda 1-------------------------------------------
ventas = Counter(tienda['Producto'])
mas_vendido = max(ventas, key=ventas.get)
print("Producto más vendido en la tienda 1:", mas_vendido)
print("Cantidad vendida:", ventas[mas_vendido])

ventas = Counter(tienda['Producto'])
menos_vendido = min(ventas, key=ventas.get)
print("Producto menos vendido en la tienda 1:", menos_vendido)
print("Cantidad vendida:", ventas[menos_vendido])

#----------Grafica Producto más vendido vs menos vendido en Tienda 1------------
productos = [mas_vendido, menos_vendido]
cantidades = [ventas[mas_vendido], ventas[menos_vendido]]

# Invertir el orden para que el más vendido aparezca primero (arriba en la horizontal)
productos = productos[::-1]
cantidades = cantidades[::-1]

# Crear gráfica de barras horizontal
plt.barh(productos, cantidades, color=['mediumseagreen', 'indianred'])

# Etiquetas de los valores en las barras
for i, valor in enumerate(cantidades):
    plt.text(valor + 0.1, i, str(valor), va='center')

# Título y etiquetas
plt.title('Producto más vs menos vendido - Tienda 1')
plt.xlabel('Cantidad Vendida')
plt.tight_layout()

# Mostrar gráfica
plt.show()
```

# 5. Envío promedio por tienda
